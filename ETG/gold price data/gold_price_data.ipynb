```python
import pandas as pd
import requests as req
from bs4 import BeautifulSoup
```

```python
url=["https://www.livechennai.com/get_goldrate_history.asp?monthno=1&yearno=",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=2&yearno=",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=3&yearno=",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=4&yearno=",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=5&yearno=",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=5&yearno=",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=7&yearno=",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=8&yearno=",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=9&yearno=",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=10&yearno=",
       "https://www.livechennai.com/get_goldrate_history.asp?monthno=11&yearno=",
       "https://www.livechennai.com/get_goldrate_history.asp?monthno=12&yearno="
    ]
url_list = []
for year in range(2010, 2024):
    for link in url:
        new_url = link+str(year)
        url_list.append(new_url)
        
```

```python
data = pd.DataFrame(url_list, columns = ['web url'])
data.to_csv("gold_data_2010-2023")
```

```python
data = pd.read_csv('gold_data_2010-2023')
data.info()
```

```python
url = data['web url'].head(167)
url
```

```python
pages=[]
for page in url:
    pages.append(req.get(page))

soups=[]
for page in pages:
    soups.append(BeautifulSoup(page.text,'lxml'))
    
    
tr_tags=[]
for soup in soups:
    tr_tags.append(soup.find_all('tr'))
```

```python

def get_monthly_data(trlist):
    month_data=[]
    for month in trlist:
          for tr in month:
            td_tag=tr.find_all('td')
            row = [i.text.strip() for i in td_tag]
            month_data.append(row)
    return month_data

```

```python
monthly_data = get_monthly_data(tr_tags)
```

```python
string='Date'
for lst in monthly_data:
  length=len(lst)
  if length < 3 or string in lst:
    monthly_data.remove(lst)
monthly_data
```

```python
count = 0
for i in monthly_data:
    if len(i) != 3:
        count += 1
print(count)
        
```

```python
def process(data):
    return [lst[0] for lst in data],[lst[1] for lst in data],[lst[2] for lst in data]
```

```python
date,gold_24k,gold_22k=process(monthly_data)
```

```python
gold_data_dict={
      "DATE":date,
    "GOLD_24k_PRICE":gold_24k,
    "GOLD_22k_PRICE":gold_22k
  }
```

```python
df =pd.DataFrame(gold_data_dict)
```
